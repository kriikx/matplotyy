```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [100, 150, 130, 180, 220]
```

```python
plt.plot(months,sales,marker="*"
            ,linestyle="dashed"
            ,linewidth=2
            ,color="red"
            ,markerfacecolor="yellow"
            ,label="Sales")
plt.legend()
plt.title("sales",fontsize=30,fontweight=30)
plt.grid(True)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.figure(figsize=(1000,1000))


plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    "Student": ["A", "B", "C"],
    "Marks": [80, 90, 70]
}

df = pd.DataFrame(data)

# Figure size
plt.figure(figsize=(7, 5))

# Theme
sns.set_style("whitegrid")

# Bar plot
sns.barplot(
    x="Student",
    y="Marks",
    data=df,
    palette="Set2",
    width=0.6
)

# Title and labels
plt.title("Student Marks", fontsize=16, fontweight="bold")
plt.xlabel("Student Name", fontsize=12)
plt.ylabel("Marks", fontsize=12)

# Y-axis limit
plt.ylim(0, 100)

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Add value labels on bars
for i, value in enumerate(df["Marks"]):
    plt.text(i, value + 1, str(value), ha="center", fontsize=11)

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Gender": ["Male", "Female", "Male", "Male", "Female", "Female", "Male"]
}

df = pd.DataFrame(data)

plt.figure(figsize=(6,4))

sns.set_style("whitegrid")

ax = sns.countplot(
    x="Gender",
    data=df,
    palette="pastel"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Number of Students")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Marks": [45,50,55,60,60,65,70,75,80,85,90,95]
}

df = pd.DataFrame(data)

plt.figure(figsize=(7,5))

sns.set_style("whitegrid")

sns.kdeplot(
    x="Marks",
    data=df,
    fill=True,
    color="skyblue",
    linewidth=2
)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Density")

plt.show()
```
