```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [100, 150, 130, 180, 220]
```

```python
plt.plot(months,sales,marker="*"
            ,linestyle="dashed"
            ,linewidth=2
            ,color="red"
            ,markerfacecolor="yellow"
            ,label="Sales")
plt.legend()
plt.title("sales",fontsize=30,fontweight=30)
plt.grid(True)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.figure(figsize=(1000,1000))


plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr"]
sales = [100, 120, 150, 180]

plt.plot(months, sales, label="Sales",color="green",marker="o")

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.tight_layout()
plt.savefig("graph.png")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [40, 30, 20, 10]
labels = ["Food", "Rent", "Travel", "Others"]
colors=["cyan","yellow","green","blue"]

plt.pie(data, labels=labels,
        autopct="%1.1f%%",colors=colors)
plt.title("Agrigature biased data",fontsize=13,color="green")

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [40, 30, 20, 10]
labels = ["Food", "Rent", "Travel", "Others"]

explode = [0, 0, 0, 0.3]

plt.pie(data,
        labels=labels,
        explode=explode,
        shadow=True,startangle=90)

plt.show()
```

```python
import matplotlib.pyplot as plt

expenses = [40, 30, 20, 10]

labels = ["Food", "Rent", "Travel", "Others"]

colors = ["red", "blue", "green", "orange"]

explode = [0.05, 0, 0, 0]

plt.figure(figsize=(6,6))

plt.pie(expenses,
        labels=labels,
        colors=colors,
        autopct="%1.1f%%",
        explode=explode,
        shadow=True,
        startangle=90)

plt.title("Monthly Expenses")

plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5]
marks = [40, 55, 60, 75, 90]

plt.figure(figsize=(7,5))

plt.scatter(hours,
            marks,
            color="blue",
            s=120,
            marker="o",
            edgecolors="black",
            alpha=0.8,
            label="Students")

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")


plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 62, 65, 70, 72, 75, 80, 82, 90]

plt.figure(figsize=(7,5))

plt.hist(
    marks,
    bins=5,
    color="orange",
    edgecolor="black",
    linewidth=1,
    alpha=0.8
)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")


plt.show()
```

```python
import matplotlib.pyplot as plt

# First graph
plt.subplot(1, 2, 1)

plt.plot([1, 2, 3], [4, 5, 6],color="green")
plt.title("Line")

# Second graph
plt.subplot(1, 2, 2)

plt.bar(["A", "B", "C"], [5, 3, 7],color="red")
plt.title("Bar")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))

plt.subplot(2,2,1)
plt.plot([1,2,3],[2,4,6],color="green")
plt.title("Line Plot")

plt.subplot(2,2,2)
plt.bar(["A","B","C"],[3,5,2],color="pink")
plt.title("Bar Graph")

plt.subplot(2,2,3)
plt.scatter([1,2,3],[4,2,5])
plt.title("Scatter Plot")

plt.subplot(2,2,4)
plt.hist([45,50,55,60,65,70,75,80],bins=4,edgecolor=
"black",color="red")
plt.title("Histogram")

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Watch"],
    "Sales": [50, 80, 30, 20]
}

df = pd.DataFrame(data)

plt.figure(figsize=(7,5))

plt.bar(df["Product"], df["Sales"], color="green")

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.style.use("ggplot")


plt.show()
```
